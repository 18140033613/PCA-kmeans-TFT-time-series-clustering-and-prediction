```python
#We use the autogluon library to complete tasks
import pandas as pd
from autogluon.timeseries import TimeSeriesDataFrame, TimeSeriesPredictor
df = pd.read_csv("E:/TFT-Prediction/TFT-Prediction/sample_data/sample_time_varying.csv") #You may modify it to your path
static_features_df = pd.read_csv("E:/TFT-Prediction/TFT-Prediction/sample_data/sample_static.csv")
data = TimeSeriesDataFrame.from_data_frame(
    df,
    id_column="item_id",  
    timestamp_column="timestamp",  
    static_features_df=static_features_df,
)
data.head()
```

```python
prediction_length=20  #The length of time steps required for prediction
train_data, test_data = data.train_test_split(prediction_length)

predictor = TimeSeriesPredictor(
    prediction_length=prediction_length, 
    eval_metric="MASE",
    path="Autogluon-VVchronos_tiny",  #output path
    target="VV"   #Column names that require prediction 
)
predictor.fit(
    train_data, 
    verbosity=4,
    hyperparameters={
        "Chronos": {         
            "model_path": "tiny",  #or "large", "base","medium","mini",
            "batch_size": 64,
            "device": "cuda",  #or "cpu" if your cuda is not available
        }
    },
)
```

```python
%%time
predictions = predictor.predict(train_data)
```

```python
predictor.evaluate(test_data)
```

```python
predictor.plot(
    data=data, 
    predictions=predictions, 
    item_ids=["id_582", "id_7700",  "id_11358"],  #Specify the item_id that needs to be visualized                
    max_history_length=210,  #The length of historical time steps that need to be reviewed. My data has 209 time steps, and you can make changes based on your own data
)
```
